```python
def get_addr(x, y):
    get_code = requests.get('http://api.vworld.kr/req/address?service=address&request=getAddress&version=2.0&crs=epsg:4326&point='+x+','+y+'&format=json&type=both&zipcode=true&simple=false&key=DF50E610-13D3-392C-99C6-D4A1E1DE8512')
    json_obj = get_code.json()
    if json_obj['response']['status'] == 'NOT_FOUND':
        result = "NaN"
    else:
        result = json_obj['response']['result'][0]['text']
    return result

# API 키값
# 9834BFA9-DBB5-342F-BCB9-6E68109DCA6C
# DF50E610-13D3-392C-99C6-D4A1E1DE8512
```

```python
List = []

for i in range(0, len(df['X좌표'])):
    List.append(get_addr(str(df['X좌표'][i]),str(df['Y좌표'][i])))
```
