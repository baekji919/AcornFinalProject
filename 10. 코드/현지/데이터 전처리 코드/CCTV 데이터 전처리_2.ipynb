```python
import pandas as pd
```

#### 2017 - 2019 병합 함수화

```python
def result(name):
     return pd.concat([name[name['설치년월'] < '2018-01'],name[name['설치년월'] < '2019-01'],name[name['설치년월'] < '2020-01']], 
                      keys=['2017', '2018', '2019'])
```
