```python
import pandas as pd
```

#### 파일 불러오기

```python
light_df = pd.read_csv("C:/AcornFinalProject/2. 데이터 파일/가로등/raw/서울시 도로시설물 통계[2015-2019].csv", encoding='ms949')

light_df
```

#### 기간, 구분, 가로등 열만 추출

```python
light_df = light_df[['기간', '구분', '가로등']]

light_df
```

#### 2017년 ~ 2019년만 추출하기

```python
light_df1 = light_df[(light_df['기간'] == '2017')|(light_df['기간'] == '2018')|(light_df['기간'] == '2019')]

light_df1
```

#### 합계, 안전총괄본부, 동부도로사업소, 서부도로사업소, 남부도로사업소, 북부도로사업소, 성동도로사업소, 강서도로사업소 행 삭제

```python
list = ['합계', '안전총괄본부', '동부도로사업소', '서부도로사업소', '남부도로사업소',
        '북부도로사업소', '성동도로사업소', '강서도로사업소']
```

```python
for i in range(len(list)):
    light_df1 = light_df1.drop(light_df1[light_df1['구분'] == list[i]].index)
    light_df1
    
light_df1
```

#### 정제 데이터 저장

```python
light_df1.to_csv("C:/AcornFinalProject/2. 데이터 파일/가로등/data/서울시 도로시설물 통계.csv",
                 index=False, encoding='ms949')
```
