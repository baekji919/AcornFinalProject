### 2. NaN 전처리
#### 1) 상세영업상태명

```python
# 상세영업상태명 Nan값 처리 : 삭제
## 주소와 사업자명만 다른 동일 장소가 이미 존재
df = df[df['상세영업상태명'].notnull()].reset_index(drop=True)
df
```

```python
df.isnull().sum()
```

#### 2. 폐업일자 NaN값 처리

```python
# 컬럼 타입 변환 : 폐업일자에 맞추기
df['최종수정일자'] = df['최종수정일자'].astype(float)
```

```python
df.dtypes
```

```python
# df 분리
## open_null_df : 영업중이고 폐업일자가 null인 영화관
## closed_null_df : 영업 종료했고 폐업일자가 null인 영화관
## notnull_df : 폐업일자가 null이 아닌 영화관

open_null_df = df[(df['상세영업상태명']=='영업중')&(df['폐업일자'].isna())].reset_index(drop=True)
closed_null_df = df[(df['상세영업상태명']!='영업중')&(df['폐업일자'].isna())].reset_index(drop=True)
notnull_df = df[df['폐업일자'].notnull()].reset_index(drop=True)
notnull_df['최종수정일자'] = notnull_df['최종수정일자'].astype('int64')
print(len(open_null_df), len(closed_null_df), len(notnull_df))
```

```python
# 영업 종료했고 폐업일자가 null인 영화관의 폐업일자를 데이터 최종수정일자로 채우기
closed_null_list = (closed_null_df['최종수정일자']//1000000).values.tolist()
closed_null_list
```

```python
closed_null_df['폐업일자'] = closed_null_list
closed_null_df['최종수정일자'] = closed_null_df['최종수정일자'].astype('int64')
closed_null_df
```

```python
# 영업중이고 폐업일자가 null인 영화관의 폐업일자를 0.0으로 채우기
open_null_df['폐업일자'].fillna(0.0, inplace=True)
open_null_df['최종수정일자'] = open_null_df['최종수정일자'].astype('int64')
open_null_df
```

```python
# 세 df 합치기
result_df = pd.concat([open_null_df, closed_null_df, notnull_df], axis=0).reset_index(drop=True)
result_df
```

```python
# 컬럼 타입 변환 : 폐업일자를 int64로 변환
result_df = result_df.astype({'폐업일자' : 'int64'})
result_df
```

```python
result_df.dtypes
```

```python
result_df.isna().sum()
```

### 5. 2017년 이후 데이터 전처리
#### 1) 컬럼명 변경

```python
data = pd.read_csv('data/영화관/서울특별시 영화상영관 인허가 정보(2017~2019, cp949).csv', encoding='cp949')
data
```

```python
data.isna().sum()
```

```python
data = data.rename({'최초등록시점' : '설립일자'}, axis=1)
data
```

#### 2) 같은 영화관 row 합치기

```python
# 설립일자 기준 내림차순
data.sort_values(by=['설립일자'], axis=0, inplace=True)
data
```

```python
# 지번주소 분리 : 시, 구, 동, 번지, 건물
data['시'] = data.지번주소.str.split(' ').str[0]
data['구'] = data.지번주소.str.split(' ').str[1]
data['동'] = data.지번주소.str.split(' ').str[2]
data['번지'] = data.지번주소.str.split(' ').str[3]
data['건물'] = data.지번주소.str.split(' ').str[4:]
data
```

```python
# 사업장명 분리 : 극장, 관
data.loc[:, "new_사업장명"] = data["사업장명"].str.replace('\(|\)|1관|2관|3관|4관|5관|6관|7관|8관|9관|10관|11관|12관|제|주식회사|<<|>>|125관', " ").str.strip()
data['극장'] = data.new_사업장명.str.split(' ').str[0]
data['관'] = data.new_사업장명.str.split(' ').str[1:]
data
```

```python
# 주소 기준 같은 극장끼리 합치기
dup_data = data.drop_duplicates(['시', '구', '동', '번지', '극장', '상세영업상태명'], keep='first')
dup_data
```

```python
# 주소 기준 같은 주소끼리 합치기
dup_data2 = data.drop_duplicates(['시', '구', '동', '번지', '상세영업상태명'], keep='first')
dup_data2
```

```python
dup_data2.to_csv('data/영화관sub.csv', encoding='cp949', index=False)
```
