```python
cluster_0_list = []

for i in range(len(kmeans_list[0])) :
    xy_df = file[file.index == kmeans_list[0][i]][['X좌표(경도)', 'Y좌표(위도)']]
    cluster_0_list.append(df[(df['X좌표(경도)'].values == xy_df['X좌표(경도)'].values) & (df['Y좌표(위도)'].values == xy_df['Y좌표(위도)'].values)].index)
```

```python
cluster_0_df
```

```python
def clusterDataFrame(num) :
    listname = 'cluster_' + str(num) + '_list'
    listname = []

    for i in range(len(kmeans_list[num])) :
        xy_df = file[file.index == kmeans_list[num][i]][['X좌표(경도)', 'Y좌표(위도)']]
        listname.append(df[(df['X좌표(경도)'].values == xy_df['X좌표(경도)'].values) & (df['Y좌표(위도)'].values == xy_df['Y좌표(위도)'].values)].index)
        
    cluste_df_list_name = 'cluster_' + str(num) + '_df_list'
    cluste_df_list_name = []
    for i in range(len(listname)) :
        cluste_df_list_name.append(np.array(df.loc[listname[i][0]].tolist()))
        
    col_name = df.columns
    cluster_df_name = 'cluster_' + str(num) + '_df'
    cluster_df_name = pd.DataFrame(cluste_df_list_name, columns = col_name)
    cluster_df_name['cluster'] = num
    
    return cluster_df_name
```

```python
for i in range(9) :
    clusterDataFrame(i)
```

```python
def cluster_df_concat(num, person) :
    for i in range(num-1) :
        if i == 0 :
            cluster_df = pd.concat([clusterDataFrame(i), clusterDataFrame(i+1)], axis=0).reset_index(drop=True)
        else :
            cluster_df = pd.concat([cluster_df, clusterDataFrame(i+1)], axis=0).reset_index(drop=True)
            
    path = 'data/KMeans/cluster/' + str(person)+ '인 가구 cluster.csv'
    cluster_df.to_csv(str(path), encoding='cp949', index=False)
    
    return cluster_df
```

```python
cluster_df_concat(9, 4)
```

### 위경도별 cluster, 총점

```python
cluster_default = data[['X좌표(경도)', 'Y좌표(위도)', 'cluster']]
cluster_default
```

```python
cluster_1_person = pd.read_csv('data/KMeans/score/1인 가구 cluster별 총점.csv', encoding='cp949')
cluster_2_person = pd.read_csv('data/KMeans/score/2인 가구 cluster별 총점.csv', encoding='cp949')
cluster_4_person = pd.read_csv('data/KMeans/score/4인 가구 cluster별 총점.csv', encoding='cp949')
```

```python
# 각 가구별 cluster 총점
cluster_1_score = cluster_1_person[['cluster', '총점']]
cluster_2_score = cluster_2_person[['cluster', '총점']]
cluster_4_score = cluster_4_person[['cluster', '총점']]

score_1_person = pd.merge(cluster_default, cluster_1_score, on='cluster')
score_2_person = pd.merge(cluster_default, cluster_2_score, on='cluster')
score_4_person = pd.merge(cluster_default, cluster_4_score, on='cluster')
```
