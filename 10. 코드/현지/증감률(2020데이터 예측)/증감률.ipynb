### 2. 지하철

```python
# 파일 불러오기
file = pd.read_excel('data/증감률/지하철.xlsx', index_col=[0, 1])
file
```

```python
file.reset_index(inplace=True)
file.drop('level_1', axis=1, inplace=True)
file.rename({'level_0':'년'}, axis=1, inplace=True)
file
```

```python
file['구'] = file['주소'].str.split(' ').str[1]
file
```

```python
# 데이터 분리
data_2017 = file[file['년']==2017].reset_index(drop=True)
data_2018 = file[file['년']==2018].reset_index(drop=True)
data_2019 = file[file['년']==2019].reset_index(drop=True)
```

```python
gu = file['구'].unique()
line = file['호선'].unique()

line_df = pd.DataFrame(line, columns=['호선'])
gu_df = pd.DataFrame(gu, columns=['구'])
```

```python
line_df["key"] = 1
gu_df["key"] = 1

idx = pd.merge(line_df, gu_df, on ='key').drop("key", 1)
```

```python
line_2017=[]

for i in range(len(line)) :
    df = data_2017[data_2017['호선'] == line[i]]
    for j in range(len(gu)) :
        line_2017.append(len(df[(df['구']==gu[j])]))

len(line_2017)
```

```python
line_2018=[]
gu = idx['구'].unique()
line = idx['호선'].unique()

for i in range(len(line)) :
    df = data_2018[data_2018['호선'] == line[i]]
    for j in range(len(gu)) :
        line_2018.append(len(df[(df['구']==gu[j])]))

line_2018
```

```python
line_2019=[]
gu = idx['구'].unique()
line = idx['호선'].unique()

for i in range(len(line)) :
    df = data_2019[data_2019['호선'] == line[i]]
    for j in range(len(gu)) :
        line_2019.append(len(df[(df['구']==gu[j])]))

line_2019
```

```python
subway_df = pd.DataFrame([ x for x in zip(line_2017, line_2018, line_2019)]).reset_index(drop=True)

subway_df.rename({0:'2017년',
                1:'2018년',
                 2:'2019년'}, axis=1, inplace=True)
subway_df
```

```python
# 18-17, 19-18
subway_df['2017년 대비 2018년 증감률']=round(((subway_df['2018년'] - subway_df['2017년']) / subway_df['2018년'])*100,2)
subway_df['2018년 대비 2019년 증감률']=round(((subway_df['2019년'] - subway_df['2018년']) / subway_df['2019년'])*100,2)
subway_df.fillna(0.0, inplace=True)
subway_df
```

```python
subway = pd.concat([idx, subway_df], axis=1)
subway['데이터'] = '지하철역'
subway = subway[['데이터', '호선', '구', '2017년 대비 2018년 증감률', '2018년 대비 2019년 증감률']]
subway
```
