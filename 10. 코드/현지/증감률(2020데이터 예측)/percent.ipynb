### 1. 관서별 5대범죄 발생/검거 현황

```python
# 파일 불러오기
file = pd.read_csv('data/범죄/전체/관서별 5대범죄 발생 현황.csv', encoding='cp949')
file
```

```python
# 증감률 구하기
office = list(file['관서'].unique())
crime = list(file['죄종'].unique())

per_list=[]
office_list = []
crime_list = []

for i in range(len(office)) :
    for j in range(len(crime)) :
        df = file[(file['관서']==office[i]) & (file['죄종']==crime[j])].reset_index(drop=True)
        crime_list.append(((df['건수'][len(df['건수'])-1]-df['건수'][0]))/sum(df['건수']))
```

```python
crime_list
```

```python
# df로 생성
crime_df = pd.DataFrame(crime_list).reset_index(drop=True)
crime_df.rename({0:'percent'}, axis=1, inplace=True)
crime_df
```

```python
# 2019년 df 생성
data_2019 = file[file['년']==2019].reset_index(drop=True)
data_2019
```

```python
# 증감률 반영한 2020년 건수 리스트 생성
list_2020=[]
for i in range(len(data_2019)) :
    list_2020.append(int(data_2019['건수'][i]-crime_df['percent'][i]))
    
len(list_2020)
```

```python
# df 생성
data_2020 = data_2019.copy()
data_2020['년'] = 2020
data_2020['건수'] = list_2020
data_2020
```

```python
# 데이터 저장
data_2020.to_csv('data/FIN/범죄/관서별 5대범죄 발생 현황 2020.csv', encoding='cp949', index=False)
```

### 3. 서울시 5대 범죄 발생현황 연도별 통계

```python
# 파일 불러오기
file = pd.read_csv('data/범죄/전체/서울시 5대 범죄 발생현황 연도별 통계.csv', encoding='cp949')
file
```

```python
# 증감률 구하기
office = list(file['자치구'].unique())

list1 = []
list2 = []
list3 = []
list4 = []
list5 = []
list6 = []
list7 = []
list8 = []
list9 = []
list10 = []


for i in range(len(office)) :
    df = file[(file['자치구']==office[i])].reset_index(drop=True)
    list1.append(((df['살인_발생'][len(df['살인_발생'])-1]-df['살인_발생'][0]))/sum(df['살인_발생']))
    list2.append(((df['살인_검거'][len(df['살인_검거'])-1]-df['살인_검거'][0]))/sum(df['살인_검거']))
    list3.append(((df['강도_발생'][len(df['강도_발생'])-1]-df['강도_발생'][0]))/sum(df['강도_발생']))
    list4.append(((df['강도_검거'][len(df['강도_검거'])-1]-df['강도_검거'][0]))/sum(df['강도_검거']))
    list5.append(((df['강간강제추행_발생'][len(df['강간강제추행_발생'])-1]-df['강간강제추행_발생'][0]))/sum(df['강간강제추행_발생']))
    list6.append(((df['강간강제추행_검거'][len(df['강간강제추행_검거'])-1]-df['강간강제추행_검거'][0]))/sum(df['강간강제추행_검거']))
    list7.append(((df['절도_발생'][len(df['절도_발생'])-1]-df['절도_발생'][0]))/sum(df['절도_발생']))
    list8.append(((df['절도_검거'][len(df['절도_검거'])-1]-df['절도_검거'][0]))/sum(df['절도_검거']))
    list9.append(((df['폭력_발생'][len(df['폭력_발생'])-1]-df['폭력_발생'][0]))/sum(df['폭력_발생']))
    list10.append(((df['폭력_검거'][len(df['폭력_검거'])-1]-df['폭력_검거'][0]))/sum(df['폭력_검거']))
```

```python
# df로 생성
crime_df = pd.DataFrame([ x for x in zip(list1,list2,list3,list4,list5,list6,list7,list8,list9,list10)]).reset_index(drop=True)
crime_df.rename({0:'살인_발생',
                1:'살인_검거',
                2:'강도_발생',
                3:'강도_검거',
                4:'강간강제추행_발생',
                5:'강간강제추행_검거',
                6:'절도_발생',
                7:'절도_검거',
                8:'폭력_발생',
                9:'폭력_검거'}, axis=1, inplace=True)
crime_df
```

### 4. 서울시 청소년 5대범죄

```python
# 파일 불러오기
file = pd.read_csv('data/범죄/전체/서울시 청소년 5대범죄.csv', encoding='cp949')
file
```

```python
# 증감률 구하기
office = list(file['관서'].unique())

list1 = []
list2 = []
list3 = []
list4 = []
list5 = []


for i in range(len(office)) :
    df = file[(file['관서']==office[i])].reset_index(drop=True)
    list1.append(((df['살인'][len(df['살인'])-1]-df['살인'][0]))/sum(df['살인']))
    list2.append(((df['강도'][len(df['강도'])-1]-df['강도'][0]))/sum(df['강도']))
    list3.append(((df['강간·추행'][len(df['강간·추행'])-1]-df['강간·추행'][0]))/sum(df['강간·추행']))
    list4.append(((df['절도'][len(df['절도'])-1]-df['절도'][0]))/sum(df['절도']))
    list5.append(((df['폭력'][len(df['폭력'])-1]-df['폭력'][0]))/sum(df['폭력']))
```

```python
# df로 생성
crime_df = pd.DataFrame([ x for x in zip(list1,list2,list3,list4,list5)]).reset_index(drop=True)
crime_df.rename({0:'살인',
                1:'강도',
                2:'강간·추행',
                3:'절도',
                4:'폭력'}, axis=1, inplace=True)
crime_df.fillna(0, inplace=True)
crime_df
```

```python
# 2019년 df 생성
data_2019 = file[file['년']==2019].reset_index(drop=True)
data_2019
```

```python
# 증감률 반영한 2020년 건수 리스트 생성
list_2020_1=[]
list_2020_2=[]
list_2020_3=[]
list_2020_4=[]
list_2020_5=[]


for i in range(len(data_2019)) :
    list_2020_1.append(int(data_2019['살인'][i]-crime_df['살인'][i]))
    list_2020_2.append(int(data_2019['강도'][i]-crime_df['강도'][i]))
    list_2020_3.append(int(data_2019['강간·추행'][i]-crime_df['강간·추행'][i]))
    list_2020_4.append(int(data_2019['절도'][i]-crime_df['절도'][i]))
    list_2020_5.append(int(data_2019['폭력'][i]-crime_df['폭력'][i]))
    
len(list_2020_1)
```

```python
# df 생성
data_2020 = data_2019.copy()
data_2020['년'] = 2020

data_2020['살인'] = list_2020_1
data_2020['강도'] = list_2020_2
data_2020['강간·추행'] = list_2020_3
data_2020['절도'] = list_2020_4
data_2020['폭력'] = list_2020_5

data_2020
```

```python
# 데이터 저장
data_2020.to_csv('data/FIN/범죄/서울시 청소년 5대범죄 2020.csv', encoding='cp949', index=False)
```
