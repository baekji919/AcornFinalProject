### 1. 그리드 df 생성(400 * 400)

```python
xy_list = []

for i in range(0, 92):  
    for j in range(0, 76):
        lang = 0.0059*i + 126.73
        lat = 0.0041*j + 37.41
        xy_list.append([lang, lat])
        
xy_df = pd.DataFrame(xy_list)
xy_df.rename(columns={0:'X좌표(경도)', 1:'Y좌표(위도)'}, inplace=True)
xy_df

x_df = pd.DataFrame(xy_df['X좌표(경도)'].unique(), columns=['X좌표(경도)'])
y_df = pd.DataFrame(xy_df['Y좌표(위도)'].unique(), columns=['Y좌표(위도)'])
```
