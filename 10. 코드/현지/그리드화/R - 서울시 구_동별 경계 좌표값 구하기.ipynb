# 시 경계선 좌표

```python
# 경로 설정
setwd('C:/Users/HYEONJI/Desktop/SIG_202005')


# 패키지 호출(없을 경우 install.packages())
library(sp)
library(rgdal)
library(dplyr)
library(ggplot2)


# proj 인자 설정
from_crs <- CRS('+proj=tmerc +lat_0=38 +lon_0=127.5 +k=0.9996 +x_0=1000000 +y_0=2000000')
to_crs <- CRS('+proj=longlat +ellps=WGS84 +datum=WGS84 +no_defs')


# 시도 shp file 불러오기
shp_sido_grs <- rgdal::readOGR('SIG.shp')


# 좌표계 변환
shp_sido <- spTransform(shp_sido_grs, to_crs)
summary(shp_sido)

slotNames(shp_sido)

# shp 속성 확인
shp_sido@data
slotNames(shp_sido@data)

# 작업할 데이터 생성
data_sido <- shp_sido@data; str(data_sido)


# 기준이 되는 17개 시도명 데이터 생성
data_result <- shp_sido@data
str(data_result)
head(data_result)


# shp 파일을 data.frame으로 변경
shp_result <- fortify(shp_sido)
str(shp_result)
head(shp_result)


# data_result에서 시도명 추출 및 기준 id 생성
Name <- cbind(id=row.names(data_result), sidoname=as.character(data_result$SIG_KOR_NM))
str(Name)
head(Name)


# list를 data.frame으로 변경 후 id를 factor -> char로 변경
x <- as.data.frame(Name)
x$id <- as.character(x$id)
str(x)
head(x)


# 마무리
result <- left_join(shp_result, x)
head(result)


# csv로 저장
write.csv(result, file='서울시 좌표.csv', row.names=F)
```
