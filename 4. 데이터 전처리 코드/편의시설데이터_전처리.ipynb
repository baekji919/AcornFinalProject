#### 업태구분명 기타휴게음식점에서 편의점으로 변경

```python
etc_df2 = pd.read_csv('C:/AcornFinalProject/2. 데이터 파일/편의시설/음식점/data/기타 휴게음식점.csv', encoding='cp949')
etc_df2
```

```python
# 해당 단어 포함하는 사업장명 인덱스 리스트에 넣기

conv_df = etc_df2[etc_df2['사업장명'].str.contains('이마트|세븐일레븐|씨유|미니스톱|지에스|CU|GS|gs|cu')].index

len(conv_df)
```

```python
# 업태구분명 편의점으로 변경

for i in range(len(conv_df)):
    food_df['업태구분명'][conv_df[i]] = '편의점'
    
food_df
```

```python
# 바뀐거 확인

for i in range(len(conv_df)):
    print(food_df['업태구분명'][conv_df[i]])
    print(i)
```
