#### 데이터 전처리 함수화

```python
def cctv(name):
    cctv_df = pd.read_csv('C:/AcornFinalProject/2. 데이터 파일/CCTV/raw/서울특별시_{}_CCTV.csv'.format(name),encoding='ms949')
    cctv_df['설치년월'] = cctv_df['설치년월'].fillna(cctv_df['데이터기준일자'][0])
    
    cctv_df1 = cctv_df[['소재지도로명주소', '소재지지번주소', '설치목적구분', '카메라대수', '설치년월']]
    cctv_df1.to_csv('C:/AcornFinalProject/2. 데이터 파일/CCTV/data/서울특별시_{}_CCTV_data.csv'.format(name),index=False, encoding='ms949')
```
