```python
import numpy as np
import pandas as pd
from datetime import timedelta
import glob
import os

import warnings
warnings.filterwarnings(action='ignore')
```

```python
data = pd.read_csv('raw/서울시 대학 및 전문대학 DB 정보.csv', encoding='cp949')
data
```

```python
# 필요없는 컬럼 삭제
data.drop(['키값', '본분교', '우편번호', '전화번호', '팩스번호', '홈페이지'], axis=1, inplace=True)
data
```

```python
data.isna().sum()
```

```python
# 도로명주소, 연도 삭제
data.drop(['도로명주소', '연도'], axis=1, inplace=True)
data
```

```python
data['학교상태'].unique()
```

```python
data[data['학교상태']=='폐교']
```

```python
# 폐교된 대학 삭제
idx = data[data['학교상태']=='폐교'].index
data.drop(idx, axis=0, inplace=True)
data.reset_index(drop=True, inplace=True)
data
```

```python
# 학교상태 컬럼 삭제
data.drop('학교상태', axis=1, inplace=True)
data
```

```python
# 데이터 저장
data.to_csv('data/대학교(전처리, cp949).csv', encoding='cp949', index=False)
```
