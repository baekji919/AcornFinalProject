### 학교 데이터 전체 병합

```python
# 데이터 불러오기
pre_df1 = pd.read_csv('data/FIN/어린이집_유치원 2017-2019(전처리완료, cp949).csv', encoding='cp949')
pre_df2 = pd.read_csv('data/FIN/어린이집_유치원 2020(전처리완료, cp949).csv', encoding='cp949')
school_df1 = pd.read_csv('data/FIN/서울특별시 학교 데이터 2017-2019(전처리 완료, cp949).csv', encoding='cp949')
school_df2 = pd.read_csv('data/FIN/서울특별시 학교 데이터 2020(전처리 완료, cp949).csv', encoding='cp949')
univ_df = pd.read_csv('data/FIN/대학교(전처리 완료, cp949).csv', encoding='cp949')
```

```python
# 필요없는 컬럼 삭제
pre_df1.drop(['인허가일자', '폐업일자'], axis=1, inplace=True)
pre_df2.drop(['인허가일자', '폐업일자'], axis=1, inplace=True)
school_df1.drop(['설립일자', '소재지도로명주소'], axis=1, inplace=True)
school_df2.drop(['설립일자', '소재지도로명주소'], axis=1, inplace=True)
univ_df.drop(['설립', '행정시', '행정구', '행정동'], axis=1, inplace=True)
```

```python
pre_df1.isna().sum()
```

```python
school_df1.isna().sum()
```

```python
# 컬럼명 변경
pre_df1.rename({'업태구분명' : '학교종류', '사업장명' : '학교명'}, axis='columns', inplace=True)
pre_df2.rename({'업태구분명' : '학교종류', '사업장명' : '학교명'}, axis='columns', inplace=True)
school_df1.rename({'업태구분명' : '학교종류', '소재지지번주소' : '주소'}, axis='columns', inplace=True)
school_df2.rename({'업태구분명' : '학교종류', '소재지지번주소' : '주소'}, axis='columns', inplace=True)
```

```python
pre_df1
```

```python
school_df1
```

```python
# school_df : 상세영업상태명 컬럼 추가
school_df1['상세영업상태명'] = '영업'
school_df2['상세영업상태명'] = '영업'
```

```python
# univ_df : 상세영업상태명 컬럼 추가
univ_df['상세영업상태명']='영업'

# 2017~2020년 df 생성
univ_2017 = univ_df.copy()
univ_2018 = univ_df.copy()
univ_2019 = univ_df.copy()
univ_2020 = univ_df.copy()
```

```python
# 년 컬럼 추가
univ_2017['년'] = 2017
univ_2018['년'] = 2018
univ_2019['년'] = 2019
univ_2020['년'] = 2020
```

```python
# 2017~2019 df 병합, 컬럼 정렬
univ_2017_2019 = pd.concat([univ_2017, univ_2018, univ_2019], axis=0).reset_index(drop=True)
univ_2017_2019 = univ_2017_2019[['년', '상세영업상태명', '학교종류', '학교명', '주소', '위도', '경도']]
```

```python
univ_2017_2019
```

```python
# 학교 데이터 전체 병합
data = pd.concat([pre_df1, school_df1, univ_2017_2019], axis=0).reset_index(drop=True)
data2 = pd.concat([pre_df2, school_df2, univ_2020], axis=0).reset_index(drop=True)
data
```

```python
# 년 기준 정렬
data.sort_values(by=['년'], axis=0, inplace=True)
```

```python
data
```

```python
# 데이터 저장
data.to_csv('data/FIN/학교 2017-2019(전처리 완료, cp949).csv', encoding='cp949', index=False)
data2.to_csv('data/FIN/학교 2020(전처리 완료, cp949).csv', encoding='cp949', index=False)
```
