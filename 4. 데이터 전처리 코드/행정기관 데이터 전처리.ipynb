```python
import pandas as pd
import numpy as np
```

#### 파일 불러오기

```python
org_df1 = pd.read_csv('C:/AcornFinalProject/2. 데이터 파일/행정기관/raw/행정기관_좌표.csv', encoding='ms949')
org_df1 = org_df1.drop(org_df1[org_df1['경도'] == '?'].index)

org_df1
```

```python
org_df2 = pd.read_csv('C:/AcornFinalProject/2. 데이터 파일/행정기관/raw/도로명.csv', encoding='ms949')
org_df2 = org_df2.drop(org_df2[org_df2['경도'] == '?'].index)

org_df2
```

```python
org_df3 = pd.read_csv('C:/AcornFinalProject/2. 데이터 파일/행정기관/raw/지번.csv', encoding='ms949')
org_df3 = org_df3.drop(org_df3[org_df3['경도'] == '?'].index)

org_df3
```

#### 최초 파일 행 갯수, 정제된 데이터 행 총합 서로 같은지 확인

```python
df = pd.read_csv('C:/AcornFinalProject/2. 데이터 파일/행정기관/raw/서울시 주민자치센터 정보.csv', encoding='ms949')
len(df)
```

#### 데이터 프레임 3개 합치기

```python
df = pd.concat([org_df1, org_df2, org_df3])
df
```
