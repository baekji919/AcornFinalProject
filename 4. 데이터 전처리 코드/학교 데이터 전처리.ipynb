### 2. 2017~2019, 2020년 데이터 생성

```python
import numpy as np
import pandas as pd
from datetime import timedelta
import glob
import os

import warnings
warnings.filterwarnings(action='ignore')
```

```python
data = pd.read_csv('data/학교 전처리 완료 데이터.csv')
data
```

```python
# 필요없는 컬럼 지우기
data.drop('Unnamed: 0', axis=1, inplace=True)
data
```

```python
data.dtypes
```

```python
# 설립일자 데이터 타입 변환 : str -> int
data.loc[:, "설립일자"] = data["설립일자"].str.replace('-', "").str.strip()
data['설립일자'] = data['설립일자'].astype('int64')
data.dtypes
```

```python
# 2017년 데이터 분리
data_2017 = data[data['설립일자']<20180101]
data_2017['년'] = 2017
data_2017
```

```python
# 2018년 데이터 분리
data_2018 = data[data['설립일자']<20190101]
data_2018['년'] = 2018
data_2018
```

```python
# 2019년 데이터 분리
data_2019 = data[data['설립일자']<20200101]
data_2019['년'] = 2019
data_2019
```

```python
# 2020년 데이터 분리
data_2020 = data[data['설립일자']<20210101]
data_2020['년'] = 2020
data_2020
```

```python
# 2017~2019 데이터 concat, 컬럼 정렬
result_df = pd.concat([data_2017, data_2018, data_2019], axis=0).reset_index(drop=True)
result_df = result_df[['년', '설립일자', '학교명', '소재지지번주소', '소재지도로명주소', '위도', '경도']]
result_df
```

```python
# 데이터 저장
result_df.to_csv('data/서울특별시 학교 데이터 2017-2019(전처리 완료, cp949).csv', encoding='cp949', index=False)
```

```python
# 2020년 데이터 컬럼 정렬, 저장
data_2020 = data_2020[['년', '설립일자', '학교명', '소재지지번주소', '소재지도로명주소', '위도', '경도']]
data_2020.to_csv('data/서울특별시 학교 데이터 2020(전처리 완료, cp949).csv', encoding='cp949', index=False)
data_2020
```
