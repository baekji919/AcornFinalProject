```python
import pandas as pd
```

#### 파일 불러오기

```python
pop_df = pd.read_csv("C:/AcornFinalProject/2. 데이터 파일/가구 수/raw/서울시 가구원수별 가구수(구별) 통계[2015-2019].csv", encoding='ms949')
pop_df
```

#### 2017~2019 추출

```python
pop_df1 = pop_df[(pop_df['기간'] == 2017)|(pop_df['기간'] == 2018)|(pop_df['기간'] == 2019)]

pop_df1
```

#### 합계 행 삭제

```python
pop_df2 = pop_df1.drop(pop_df1[pop_df1['구분'] == '합계'].index)

pop_df2
```

#### 정제 데이터 저장

```python
pop_df2.to_csv('C:/AcornFinalProject/2. 데이터 파일/가구 수/data/서울시 가구원수별 가구수(구별) 통계.csv',index=False, encoding='ms949')
```
