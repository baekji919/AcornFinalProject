### 전체 데이터 전처리

#### 인허가일자, 상세영업상태명, 폐업일자, 지번주소, 도로명주소, 사업장명, 업태구분명 추출

```python
df = pd.read_csv('C:/AcornFinalProject/2. 데이터 파일/편의시설/음식점/food_v0.csv', encoding='cp949')

df
```

```python
df.columns
```

```python
df1 = df[['인허가일자', '상세영업상태명', '폐업일자', '지번주소', '도로명주소', '사업장명', '업태구분명']]

df1.head(20)
```

#### 영업상태 '폐업' 중 폐업일자 'NaN'값인것

```python
df1[(df1['상세영업상태명'] == '폐업') & (df1['폐업일자'] == 'NaN')]
```

#### 폐업일자 'NaN'값 0으로 바꾸기

```python
df1['폐업일자'] = df1['폐업일자'].fillna(0)

df1
```

```python
df2 = df1.drop(df1[(df1['폐업일자'] != 0) & (df1['폐업일자'] <= 20170101)].index)

df2
```

```python
df2['폐업일자'] = df2['폐업일자'].astype('int64')

df2
```

```python
# 폐업처리까지 파일 저장
df2.to_csv('C:/AcornFinalProject/2. 데이터 파일/편의시설/음식점/data/food_v1.csv', index=False, encoding='cp949')
```

### 백화점 주소같은것 하나로 묶기

```python
dep_df = pd.read_csv('C:/AcornFinalProject/2. 데이터 파일/편의시설/음식점/data/food_v1.csv', encoding='cp949')

dep_df
```

```python
dep_df1 = dep_df[dep_df['업태구분명'] == '백화점']

dep_df1
```

```python
dep_df1 = dep_df1.sort_values(by='인허가일자')

dep_df1
```

```python
# 지번주소 분리 : 시, 구 , 동, 번지
dep_df1['시'] = dep_df1.지번주소.str.split(' ').str[0]
dep_df1['구'] = dep_df1.지번주소.str.split(' ').str[1]
dep_df1['동'] = dep_df1.지번주소.str.split(' ').str[2]
dep_df1['번지'] = dep_df1.지번주소.str.split(' ').str[3]

dep_df1
```

```python
dep_df1['번지'] = dep_df1['번지'].str.replace('-0번지', '')
dep_df1['번지'] = dep_df1['번지'].str.replace('번지', '')

dep_df1
```

```python
# 주소가 같은 백화점끼리 묶기
dep_df2 = dep_df1.drop_duplicates(['시', '구', '동', '번지'], keep='first')
dep_df2
```

```python
# 시, 구, 동, 번지 삭제
dep_df3 = dep_df2.drop(['시', '구', '동', '번지'], axis=1)
dep_df3
```

--------------
#### 도로명주소로 중복값 하나로 합치기

```python
# 도로명주소 분리 : 시, 구 , 동, 번지
dep_df3['시'] = dep_df3.도로명주소.str.split(' ').str[0]
dep_df3['구'] = dep_df3.도로명주소.str.split(' ').str[1]
dep_df3['동'] = dep_df3.도로명주소.str.split(' ').str[2]
dep_df3['번지'] = dep_df3.도로명주소.str.split(' ').str[3]

dep_df3
```

```python
# 주소가 같은 백화점끼리 묶기
dep_df4 = dep_df3.drop_duplicates(['시', '구', '동', '번지'], keep='first')
dep_df4
```

```python
dep_df4.to_csv('C:/AcornFinalProject/2. 데이터 파일/편의시설/음식점/data/서울시_백화점.csv', index=False, encoding='cp949')
```
